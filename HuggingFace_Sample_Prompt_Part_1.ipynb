```python
# load dataset from hugging face - fka/awesome-chatgpt-prompts

import pandas as pd

df = pd.read_parquet("/content/default_train_0000.parquet")
```

Import to SQL

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('data.db')

# Write the data to a sqlite table
df.to_sql('table1', conn, if_exists='replace', index=False)
```

```python
# Let's read the data from the SQL table to make sure it has been converted correctly.

query = 'SELECT * FROM table1'
df_sql = pd.read_sql_query(query, conn)

df_sql
```
