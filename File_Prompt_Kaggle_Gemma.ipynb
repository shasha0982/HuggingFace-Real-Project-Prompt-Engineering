```python
import sqlite3

# Create a connection to the database
conn = sqlite3.connect('my_database.db')

# Convert df_1 to sql and write it to the table named 'table_1'
df_1.to_sql('table_1', conn, if_exists='replace', index=False)

conn.close()
```

```python
# Create a connection to the database
conn = sqlite3.connect('my_database.db')

# Read the first table from the SQL database
df_1_from_sql = pd.read_sql('SELECT * from table_1', conn)

# Read the second table from the SQL database
df_2_from_sql = pd.read_sql('SELECT * from table_2', conn)

# Close connection
conn.close()

# Preview the first few rows of the first DataFrame
df_1_from_sql.head()
```
